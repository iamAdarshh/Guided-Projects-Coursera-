### Task 1: Data Loading and Exploration

```python
train = pd.read_csv("data/train.csv")
test = pd.read_csv("data/test.csv")
```

### Task 2: Visualizing the Target Distribution

```python
fig, ax = plt.subplots(figsize=(16, 6))
plt.subplot(1, 2, 1)
sns.distplot(train['revenue'], kde = False);
plt.title('Distribution of revenue');
plt.subplot(1, 2, 2)
sns.distplot(np.log1p(train['revenue']), kde=False)
plt.title("Distribution of log-transformed revenue");

#Its better to use log-transformed distribution of revenue since its less skewwed.
```

```python
train['log revenue'] = np.log1p(train['revenue'])
```

### Task 3: Relationship between Film Revenue and Budget

```python
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
sns.scatterplot(train["budget"], train['revenue'])
plt.title("revenue vs budget")

plt.subplot(1, 2, 2)
sns.scatterplot(np.log1p(train['budget']), train['log revenue'])
plt.title("log-revenue vs log-budget")

#relation is shown more clearly on right hand side.
#yes there is relation between budget and revenue, and budget affects revenue.
```

```python
train['log budget'] = np.log1p(train['budget'])
test['log budget'] = np.log1p(test["budget"])
```

### Task 4: Does having an Official Homepage Affect Revenue?

```python
train['homepage'].value_counts().head(10)
```

```python
train['has_homepage'] = 0
train.loc[train['homepage'].isnull() == False, 'has_homepage'] = 1
test['has_homepage'] = 0
test.loc[test['homepage'].isnull() == False, 'has_homepage'] = 1
```

### Task 5: Distribution of Languages in Film

### Task 7: Do Film Descriptions Impact Revenue?

```python
import eli5
from sklearn.linear_model import LinearRegression
```

```python
vectorizer = TfidfVectorizer(
    sublinear_tf=True,
    analyzer='word',
    token_pattern=r'\w{1,}',
    ngram_range=(1, 2),
    min_df=5
)
#min_df is for hard treshold and ignore words with frequence less than 5.
#ngarm_range(unigram, bigram), single and pairs of words occurring together.

overview_text = vectorizer.fit_transform(train['overview'].fillna(" "))
linreg = LinearRegression()
linreg.fit(overview_text, train['log revenue'])

#eli5 explains the prediction.
eli5.show_weights(linreg, vec = vectorizer, top = 20, feature_filter = lambda x: x!= '<BIAS>')

#we can clearly see that some words is clearly used predict movie revenue.
```
